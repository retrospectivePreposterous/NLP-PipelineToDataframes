```python
import transformers
import pandas as pd
```

# BETO PIPELINE

```python
# Cargar Beto pipeline
beto = transformers.pipeline('sentiment-analysis', model='finiteautomata/beto-sentiment-analysis', return_all_scores=True)
```

```python
data = pd.read_csv('tweets_prepro.csv', lineterminator='\n')
```

```python
df_plain_beto =pd.DataFrame()

df_plain_beto['tweet_prepro'] = data['tweet_prepro'][0:100]
inference = beto(df_plain_beto['tweet_prepro'].tolist())

df_plain_beto['neg_score_beto'] = [i[0]['score'] for i in inference]
df_plain_beto['neu_score_beto'] = [i[1]['score'] for i in inference]
df_plain_beto['pos_score_beto'] = [i[2]['score'] for i in inference]

df_plain_beto_scores = df_plain_beto[['neg_score_beto','neu_score_beto','pos_score_beto']]

df_plain_beto['pred_beto'] = df_plain_beto_scores.apply(lambda row: df_plain_beto_scores.columns[row.argmax()], axis=1) 
df_plain_beto.loc[df_plain_beto.pred_beto == 'neg_score_beto', 'pred_beto'] = -1
df_plain_beto.loc[df_plain_beto.pred_beto == 'neu_score_beto', 'pred_beto'] = 0
df_plain_beto.loc[df_plain_beto.pred_beto == 'pos_score_beto', 'pred_beto'] = 1
```

# NLP TOWN PIPELINE

```python
# Cargar Nlptown pipeline
nlptown = transformers.pipeline(task = 'sentiment-analysis',
                     model = 'nlptown/bert-base-multilingual-uncased-sentiment',
                     tokenizer = 'nlptown/bert-base-multilingual-uncased-sentiment',
                                return_all_scores=True)
```

```python
data = pd.read_csv('tweets_prepro.csv', lineterminator='\n')
```

```python
df_plain_nlptown =pd.DataFrame()

df_plain_nlptown['tweet_prepro'] = data['tweet_prepro'][0:100]
inference = nlptown(df_plain_nlptown['tweet_prepro'].tolist())

df_plain_nlptown['1star_score_nlptown'] = [i[0]['score'] for i in inference]
df_plain_nlptown['2star_score_nlptown'] = [i[1]['score'] for i in inference]
df_plain_nlptown['3star_score_nlptown'] = [i[2]['score'] for i in inference]
df_plain_nlptown['4star_score_nlptown'] = [i[3]['score'] for i in inference]
df_plain_nlptown['5star_score_nlptown'] = [i[4]['score'] for i in inference]



df_plain_nlptown_scores = df_plain_nlptown[['1star_score_nlptown','2star_score_nlptown','3star_score_nlptown',
                                           '4star_score_nlptown','5star_score_nlptown']]

df_plain_nlptown['pred_nlptown'] = df_plain_nlptown_scores.apply(lambda row: df_plain_nlptown_scores.columns[row.argmax()], axis=1) 
df_plain_nlptown.loc[df_plain_nlptown.pred_nlptown == '1star_score_nlptown', 'pred_nlptown'] = -1
df_plain_nlptown.loc[df_plain_nlptown.pred_nlptown == '2star_score_nlptown', 'pred_nlptown'] = -1
df_plain_nlptown.loc[df_plain_nlptown.pred_nlptown == '3star_score_nlptown', 'pred_nlptown'] = 0
df_plain_nlptown.loc[df_plain_nlptown.pred_nlptown == '4star_score_nlptown', 'pred_nlptown'] = 1
df_plain_nlptown.loc[df_plain_nlptown.pred_nlptown == '5star_score_nlptown', 'pred_nlptown'] = 1
```
